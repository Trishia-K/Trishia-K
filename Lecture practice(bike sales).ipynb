### 5.Visuals to show relationship between two continous variables

```python
#Scatter plot to show relationship between Unit_Price and Unit_Cost
BikeSales.plot.scatter(x='Unit_Price', y='Unit_Cost', figsize=(10,5), title='Unit_Price'+" VS "+ 'Unit_Cost')
```

### 6.Visuals to show relationship between two categorical variables

```python
#Generating a cross-tabulation between 'Age_Group' and 'Customer_Gender'
Crosstable=pd.crosstab(index=BikeSales['Age_Group'],columns=BikeSales['Customer_Gender'])
print(Crosstable)
```

```python
#Bar plot to show relationship between Age_Group and Customer_Gender
Crosstable.plot.bar()
```

### 7.Visuals to show relationship between one continuous variable and one categorical variable

```python
#Box plot to show relationship between Unit_Price and Age_Group.
BikeSales.boxplot(column='Unit_Price', by='Age_Group', figsize=(18,10))
```

ii.Statistical test for two categorical variables(No.6)

```python
#Chi- square test for Crosstable(Age_Group and Customer_Gender)
from scipy.stats import chi2_contingency
Chi_square_result = chi2_contingency(Crosstable)
print(f"The P-value is {Chi_square_result[1]}")
```

Since the P-value is less than 0.05,age_group and customer_gender are correlated.
