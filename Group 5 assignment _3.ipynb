##### 1.Marginal probability of any continuous variable
             Marginal probability for Applicant Income

```python
#Creating bins
bins = [0,2000,4000,6000,8000,10000,20000,30000,40000,50000]
labels = ['0-2000','2000-4000','4000-6000','6000-8000','8000-10000','10000-20000','20000-30000','30000-40000','40000-50000']

#Creating a new column for ApplicantIncome_Bin
Loan_Data['ApplicantIncome_bin'] = pd.cut(Loan_Data['ApplicantIncome'], bins = bins, labels=labels, right=False)

##Calculating the Marginal Probabilities
ApplicantIncome_probabilities = Loan_Data['ApplicantIncome_bin'].value_counts(normalize=True).sort_index()
ApplicantIncome_probabilities
```

##### 3.Probability that a "female" applicant has her Loan_Status reflecting "Y"

```python
#Filtering the data for only female applicants whose loan status is Y
filtered_data = (Loan_Data['Gender'] == 'Female') & (Loan_Data['Loan_Status'] == 'Y')

#Calculating number of outcomes from the filtered data
Female_Y_count = Loan_Data[filtered_data]['Gender'].count()
Female_Y_count

#Calculating the probability for females with a loan status reflecting Y
Female_Y_probability=Female_Y_count/len(Loan_Data)
Female_Y_probability

#Calculating the Y probability
Y_count=Loan_Data[Loan_Data['Loan_Status']=="Y"]['Loan_Status'].count()
Y_probability=Y_count/len(Loan_Data)

#Calculating the final conditional probability of females given a loan status reflecting Y
Final_probability=Female_Y_probability/Y_probability
Final_probability
print(f"The probability of females with a loan status reflecting Y is {Final_probability}")
```

##### 5.Explain the relationship between the variables "Gender" and "Loan Status"

```python
#Running the chisquare test to first find out the relationship

#Generating a cross-tabulation between 'Gender' and 'Loan Status'
CrosstabResult=pd.crosstab(index=Loan_Data['Gender'],columns=Loan_Data['Loan_Status'])
print(CrosstabResult)

from scipy.stats import chi2_contingency
ChiSqResult = chi2_contingency(CrosstabResult)
print('The P-Value is:', ChiSqResult[1])
```
